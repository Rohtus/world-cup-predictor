## Testeo la primera función

```python
def build_player_vector(events):

    # GENERO DF DE EVENTS
    player_events = []

    for event in events:

        if "player" not in event:
            continue

        player_events.append({
            "player_id": event["player"]["id"],
            "player_name": event["player"]["name"],
            "team_name": event["team"]["name"],
            "event_type": event["type"]["name"]
        })

    player_events_df = pd.DataFrame(player_events)

    # FILTRO EVENTOS ADMINISTRATIVOS
    ADMIN_EVENTS = [
        "Starting XI",
        "Half Start",
        "Half End",
        "Substitution",
        "Tactical Shift",
        "Injury Stoppage",
        "Referee Ball-Drop",
    ]

    player_events_clean = player_events_df[
        ~player_events_df["event_type"].isin(ADMIN_EVENTS)
    ].copy()

    # PLAYER VECTOR
    player_vector = (
        player_events_clean
        .pivot_table(
            index=[
                "player_id",
                "player_name",
                "team_name"
            ],
            columns="event_type",
            values="event_type",
            aggfunc="count",
            fill_value=0
        )
        .reset_index()
    )

    # TESTEO
    assert player_vector["player_id"].is_unique, "Hay jugadores duplicados"
    assert player_vector.isna().sum().sum() == 0, "Hay valores NaN"
    assert player_vector["team_name"].nunique() == 2, "No hay exactamente 2 equipos"
    assert "Pass" in player_vector.columns, "No existe la columna Pass"

    print("All tests passed")

    return player_vector
```

## Genero función para todos los partidos

```python
# Devuelve un json de cada partido

def load_events(match_id):
    event_file = EVENTS_DIR / f"{match_id}.json"

    with open(event_file, "r", encoding="utf-8") as f:
        events = json.load(f)

    return events
```

```python
def build_worldcup_player_vector(matches_df, events_dir, save_path=None):
    """
    Construye un player vector acumulado para todos los partidos de un Mundial.

    Entrada:
    - matches_df: DataFrame con columna match_id
    - events_dir: carpeta donde están los events JSON
    - save_path: ruta opcional para guardar CSV

    Salida:
    - player_vector_worldcup: 1 fila por jugador en el torneo
    - player_vectors_match_level: 1 fila por jugador por partido
    """

    all_player_vectors = []

    for match_id in matches_df["match_id"]:

        event_file = events_dir / f"{match_id}.json"

        with open(event_file, "r", encoding="utf-8") as f:
            events = json.load(f)

        player_vector = build_player_vector(events)

        player_vector["match_id"] = match_id

        all_player_vectors.append(player_vector)

    # 1 fila por jugador por partido
    player_vectors_match_level = pd.concat(
    all_player_vectors,
    ignore_index=True
)


    # Columnas identificatorias
    id_cols = [
        "player_id",
        "player_name",
        "team_name"
    ]

    # Columnas numéricas/eventos
    numeric_cols = [
    col for col in player_vectors_match_level.select_dtypes(include="number").columns
    if col not in ["match_id", "player_id"]
    ]

    # 1 fila por jugador en todo el Mundial
    player_vector_worldcup = (
        player_vectors_match_level
        .groupby(id_cols)[numeric_cols]
        .sum()
        .reset_index()
    )

    # Tests básicos
    # assert player_vector_worldcup["player_id"].is_unique, "Hay jugadores duplicados"
    assert player_vector_worldcup.isna().sum().sum() == 0, "Hay valores NaN"
    assert "Pass" in player_vector_worldcup.columns, "No existe la columna Pass"

    print("World Cup player vector created successfully")
    print("Match-level shape:", player_vectors_match_level.shape)
    print("World Cup-level shape:", player_vector_worldcup.shape)

    if save_path is not None:
        player_vector_worldcup.to_csv(save_path, index=False)
        print("Saved to:", save_path)

    return player_vector_worldcup, player_vectors_match_level
```

```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

processed_dir = BASE_DIR / "data" / "processed"
processed_dir
```

```python
player_vector_2018 = pd.read_csv(processed_dir / "player_vector_2018.csv")
player_vector_2022 = pd.read_csv(processed_dir / "player_vector_2022.csv")

player_vector_2018.shape, player_vector_2022.shape
```

```python
cols_2018 = set(player_vector_2018.columns)
cols_2022 = set(player_vector_2022.columns)

print("Solo 2018:")
print(sorted(cols_2018 - cols_2022))

print("\nSolo 2022:")
print(sorted(cols_2022 - cols_2018))
```

```python
# Unifico esquema de columnas entre 2018 y 2022

all_columns = sorted(cols_2018 | cols_2022)

player_vector_2018_common = player_vector_2018.reindex(
    columns=all_columns,
    fill_value=0
)

player_vector_2022_common = player_vector_2022.reindex(
    columns=all_columns,
    fill_value=0
)
```

```python
assert list(player_vector_2018_common.columns) == list(player_vector_2022_common.columns)

player_vector_2018_common.shape, player_vector_2022_common.shape
```

```python
player_historical_vector_v01 = pd.concat(
    [player_vector_2018_common, player_vector_2022_common],
    ignore_index=True
)

player_historical_vector_v01.shape
```
