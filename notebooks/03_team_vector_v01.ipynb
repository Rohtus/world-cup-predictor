```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

processed_dir = BASE_DIR / "data" / "processed"

player_match_2018 = pd.read_csv(
    processed_dir / "player_vectors_2018_match_level.csv"
)

player_match_2022 = pd.read_csv(
    processed_dir / "player_vectors_2022_match_level.csv"
)

player_match = pd.concat(
    [player_match_2018, player_match_2022],
    ignore_index=True
)

player_match.shape
```

```python
player_match.groupby(
    ["match_id", "team_name"]
)["player_id"].nunique().describe()
```

```python
ID_COLS = [
    "player_id",
    "player_name",
    "team_name",
    "match_id",
]

feature_cols = [
    col for col in player_match.columns
    if col not in ID_COLS
]

team_vector_v01 = (
    player_match
    .groupby(["match_id", "team_name"], as_index=False)[feature_cols]
    .sum()
)

team_vector_v01.shape
```
