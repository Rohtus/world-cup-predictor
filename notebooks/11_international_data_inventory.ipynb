# 11 - International Data Inventory

This notebook explores the international competitions available in the StatsBomb Open Data directory.

The goal is to identify which competitions can be used to build a more updated player database for the World Cup Predictor 2026 project.

# EDA

## Imports

```python
import pandas as pd
from pathlib import Path
```

## Paths

```python
BASE_DIR = Path("..")
DATA_DIR = BASE_DIR / "data"
RAW_DIR = DATA_DIR / "raw"
STATSBOMB_DATA_DIR = RAW_DIR / "statsbomb" / "data"
```

## Load Competitions

```python
competitions_path = STATSBOMB_DATA_DIR / "competitions.json"

competitions = pd.read_json(competitions_path)

competitions.head()
```

## Basic Inspections

```python
competitions.shape
```

```python
competitions.info()
```

```python
competitions.columns
```

## Clean Competitions Table

```python
competition_cols = [
    "competition_id",
    "season_id",
    "country_name",
    "competition_name",
    "competition_gender",
    "competition_youth",
    "competition_international",
    "season_name",
    "match_available",
    "match_available_360",
    "match_updated",
    "match_updated_360",
]

available_cols = [col for col in competition_cols if col in competitions.columns]

competitions_clean = competitions[available_cols].copy()

competitions_clean.head()
```

## Filter international male competitions

```python
international_male = competitions_clean[
    (competitions_clean["competition_gender"] == "male") &
    (competitions_clean["competition_international"] == True) &
    (competitions_clean["competition_youth"] == False)
].copy()

international_male.sort_values(
    by=["competition_name", "season_name"]
)
```

```python
international_male["competition_name"].value_counts()
```

## Count available international competitions

```python
international_male[
    [
        "competition_id",
        "season_id",
        "competition_name",
        "season_name",
        "country_name",
        "match_available"
    ]
].sort_values(
    by=["competition_name", "season_name"]
)
```

## Candidate competitions for V1 player database

```python
candidate_keywords = [
    "FIFA World Cup",
    "Copa America",
    "UEFA Euro",
    "European Championship",
    "Asian Cup",
    "Africa Cup",
    "CONCACAF",
    "Gold Cup",
]

pattern = "|".join(candidate_keywords)

candidate_competitions = international_male[
    international_male["competition_name"].str.contains(
        pattern,
        case=False,
        na=False
    )
].copy()

candidate_competitions.sort_values(
    by=["competition_name", "season_name"]
)
```

## Save inventory

```python
output_path = DATA_DIR / "processed" / "international_competitions_inventory.csv"

candidate_competitions.to_csv(output_path, index=False)

output_path
```
